````python
fname = "tesla_gateway_backup.bin"
raw = open(fname, 'rb').read()
print(f'Velikost souboru zalohy: {len(raw)} bytu')
````

```
0x000000-0x020000  mtd0  boot+cfg     (128 KB)   - Bootloader (unchanged)
0x020000-0x200000  mtd1  linux        (1.9 MB)   - Linux kernel
0x200000-0x420000  mtd2  rootfs       (2.1 MB)   - Root filesystem
0x420000-0x1000000 mtd3  jffs2-fs     (11.9 MB)  - User partition
```

````python
partisny = ['boot+cfg', 'linux', 'rootfs', 'jffs2-fs']
adresy = [0, 0x020000, 0x200000, 0x420000, 0x1000000]
dlt = 0x80
rozsahy = [[(a-dlt) if (a-dlt) > 0 else 0, (a+dlt) if (a+dlt) < adresy[-1] else adresy[-1]] for a in adresy]
def print_data(adr, data):
    print(f'0x{adr:07X}:', end='')
    cnt = 0
    asc = ''
    for d in data:
        print(f"{' ' if cnt % 4 == 0 else ''}{d:02X}", end='')
        asc += chr(d) if 32 <= d <= 126 else '.'
        cnt += 1
    print(f'  {asc}')
b2l = 0x20
for e, r in enumerate(rozsahy):
    for a in range((r[1] - r[0]) // b2l):
        adr = r[0] + a * b2l
        data = raw[adr:adr+b2l]
        print_data(adr, data)
    if r[1] != adresy[-1]:
        print(f'...\n{partisny[e]}\n...')
````
